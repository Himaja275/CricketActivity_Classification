```python
knn_train=[0.417,0.683,0.4]
knn_test=[0.47,0.57,0.45]
names=["KNN","Random Forest","SVM"]
```

```python
rf_train=[0.43,0.63,0.43]
rf_test=[0.5,0.5,0.47]
```

```python
combined_train=[0.43,0.65,0.57]
combined_test=[0.45,0.65,0.53]
```

```python
import seaborn as sns
import matplotlib .pyplot as plt
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

sns.barplot(x=names, y=knn_train, ax=axes[0], palette="Blues")
axes[0].set_title('Train Accuracy')
axes[0].set_xlabel('Models')
axes[0].set_ylabel('Accuracy')

# Plot the second subplot (KNN_test)
sns.barplot(x=names, y=knn_test, ax=axes[1], palette="Reds")
axes[1].set_title('Test Accuracy')
axes[1].set_xlabel('Models')
axes[1].set_ylabel('Accuracy')

# Customize the style
# plt.suptitle('SFS using KNN', fontsize=16)
# plt.tight_layout()  # Adjust spacing between subplots

# Rotate x-axis labels for better readability
axes[0].tick_params(axis='x', rotation=0)
axes[1].tick_params(axis='x', rotation=0)

# Add grid lines
axes[0].grid(axis='y', linestyle='--', alpha=0.7)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

sns.barplot(x=names, y=rf_train, ax=axes[0], palette="Blues")
axes[0].set_title('Train Accuracy')
axes[0].set_xlabel('Models')
axes[0].set_ylabel('Accuracy')

# Plot the second subplot (KNN_test)
sns.barplot(x=names, y=rf_test, ax=axes[1], palette="Reds")
axes[1].set_title('Test Accuracy')
axes[1].set_xlabel('Models')
axes[1].set_ylabel('Accuracy')

# Customize the style
# plt.suptitle('SFS using Random Forest', fontsize=16)
# plt.tight_layout()  # Adjust spacing between subplots

# Rotate x-axis labels for better readability
axes[0].tick_params(axis='x', rotation=0)
axes[1].tick_params(axis='x', rotation=0)

# Add grid lines
axes[0].grid(axis='y', linestyle='--', alpha=0.7)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)
```
