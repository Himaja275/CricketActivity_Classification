```python
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import validation_curve
import matplotlib.pyplot as plt

# Create an SVM model object
svm_model = SVC(kernel='linear', C=1)

# Define the range of C values to test
C_range = np.logspace(-3, 3, 7)

# Calculate training and validation scores for varying C values
train_scores, valid_scores = validation_curve(
    svm_model, X_train, y_train, param_name="C", param_range=C_range,
    cv=5, scoring="accuracy")

# Calculate mean and standard deviation of training and validation scores
mean_train_scores = np.mean(train_scores, axis=1)
std_train_scores = np.std(train_scores, axis=1)
mean_valid_scores = np.mean(valid_scores, axis=1)
std_valid_scores = np.std(valid_scores, axis=1)

# Plot the results
plt.figure(figsize=(10, 6))
plt.title("Validation Curve with SVM")
plt.xlabel("C")
plt.ylabel("Score")
plt.ylim(0.0, 1.1)
plt.semilogx(C_range, mean_train_scores, label="Training score", color="darkorange")
plt.fill_between(C_range, mean_train_scores - std_train_scores,
                 mean_train_scores + std_train_scores, alpha=0.2,
                 color="darkorange")
plt.semilogx(C_range, mean_valid_scores, label="Cross-validation score",
             color="navy")
plt.fill_between(C_range, mean_valid_scores - std_valid_scores,
                 mean_valid_scores + std_valid_scores, alpha=0.2,
                 color="navy")
plt.legend(loc="best")
plt.show()
```

```python
train_sizes, train_scores, test_scores = learning_curve(
    svm_model, X_train, y_train, cv=5, scoring='accuracy', 
    train_sizes=np.linspace(0.1, 1.0, 10))

train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

plt.plot(train_sizes, train_mean, label='Training accuracy')
plt.fill_between(train_sizes, train_mean - train_std,
                 train_mean + train_std, alpha=0.15)
plt.plot(train_sizes, test_mean, label='Validation accuracy')
plt.fill_between(train_sizes, test_mean - test_std,
                 test_mean + test_std, alpha=0.15)
plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
