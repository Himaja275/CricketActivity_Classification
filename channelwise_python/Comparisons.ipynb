```python
import pandas as pd
import matplotlib.pyplot as plt

# List of file names
file_names = ["kappas_data1.csv", "kappas_data2.csv", "kappas_data3.csv", "kappas_data4.csv", "kappas_data5.csv", "kappas_data6.csv", "kappas_data7.csv", "kappas_data8.csv", "kappas_data9.csv"]

# Initialize an empty list to store data from each file
data_list = []

# Loop through each file and extract the first row, first column data
for file_name in file_names:
    # Load the CSV file
    df = pd.read_csv(file_name)
    
    # Extract the value from the first row, first column
    first_value = df.iloc[0, 0]
    
    # Append the value to the data_list
    data_list.append(first_value)

# Create a plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.plot(range(1, 10), data_list, marker='o', linestyle='-')
plt.xlabel('File Number')
plt.ylabel('Value from 1st Row, 1st Column')
plt.title('Plot of 1st Row, 1st Column from Multiple CSV Files')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# List of file names
file_names = ["kappas_data1.csv", "kappas_data2.csv", "kappas_data3.csv", "kappas_data4.csv", "kappas_data5.csv", "kappas_data6.csv", "kappas_data7.csv", "kappas_data8.csv", "kappas_data9.csv"]

# Number of rows and columns in each file
num_rows = 5
num_columns = 2

# Initialize 10 empty lists
data_lists = [[] for _ in range(num_rows * num_columns)]

# Loop through each file and append the data to the corresponding list
for file_name in file_names:
    # Load the CSV file
    df = pd.read_csv(file_name)

    # Loop through rows and columns and append the corresponding data to the lists
    for row in range(num_rows):
        for col in range(num_columns):
            data = df.iloc[row, col]
            data_lists[row * num_columns + col].append(data)

# Now you have 10 lists containing the corresponding data
```

```python
plot_names = [
    "Random Forest Train", "Random Forest Test",
    "Logistic Regression Train", "Logistic Regression Test",
    "SVC Train", "SVC Test",
    "KNN Train", "KNN Test",
    "Perceptron Train", "Perceptron Test"
]

# Plot each inner list with corresponding name
for i, (inner_list, name) in enumerate(zip(data_lists, plot_names)):
    plt.figure(figsize=(8, 4))  # Adjust the figure size as needed
    plt.plot(inner_list, color='green', linestyle='-', marker='o', markersize=8)
    plt.title(name)
    plt.xlabel('Data Point Index')
    plt.ylabel('Value')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```
